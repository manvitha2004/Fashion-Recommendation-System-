```python
import numpy as np
import pandas as pd
from sklearn.decomposition import TruncatedSVD
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Load the CSV files from Google Drive
train_path = '/content/drive/MyDrive/mnist/fashion-mnist_train.csv'
test_path = '/content/drive/MyDrive/mnist/fashion-mnist_test.csv'
```

```python
df_train = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
# Creating independent variable and target variable
X = df_train.iloc[:, 1:]
y = df_train.iloc[:, 0]

test_X = df_test.iloc[:, 1:]
test_y = df_test.iloc[:, 0]
```

```python
# Feature Scaling
X = X / 255
test_X = test_X / 255
```

```python
# Reshape feature matrix into 2D array
X_flat = X.values.reshape(X.shape[0], -1)
test_X_flat = test_X.values.reshape(test_X.shape[0], -1)
```

```python
# Apply Singular Value Decomposition (SVD)
n_components = 100  # You can adjust this based on your preference
svd = TruncatedSVD(n_components=n_components, random_state=42)

X_svd = svd.fit_transform(X_flat)
test_X_svd = svd.transform(test_X_flat)
```

```python
# Split data into training and test sets
X_train_svd, X_test_svd, y_train, y_test = train_test_split(X_svd, y, test_size=0.2, random_state=1)
```

```python
# Train a classifier (Random Forest, for example)
classifier = RandomForestClassifier(n_estimators=100, random_state=42)
classifier.fit(X_train_svd, y_train)
```

```python
# Make predictions on the test set
test_set_predictions = classifier.predict(X_test_svd)
```

```python
# Evaluate the accuracy
accuracy = accuracy_score(y_test, test_set_predictions)
print(f'Accuracy using SVD: {accuracy * 100:.3f}%')
```
